```python
# this notebook in particular deals with structured data only
name = ['John', 'Jane', 'Jim', 'Pikachu', 'T-Rex']
age = [10,20,11,20,55]

print(len(name))
print(len(age))
```

```python
# as long as the length is same, we can convert it into a DATA FRAME
# DF -> rectangular (table like) representation of data

my_data_dict = {
    "name" : name,
    "age" : age
}
print(my_data_dict)
```

```python
import pandas as pd
data = pd.DataFrame(my_data_dict)
print(data)
```

```python
# DataFrames are used to analyze data -> data visualization, data cleaning, feature engineering, ML, EDA
# Slicing and dicing

# chopping in 2 levels-> ROW LEVEL and COLUMN LEVEL
# operated by either NAME of columns or INDEX of columns!

# index selection -> iloc
# name selection -> loc

# [ from_row: to_row , from_column: to_column] -> syntax with both loc and iloc

data = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/EY_batch8_11Nov_AIplusOpenAI/refs/heads/main/data.csv')
data.tail() # by deafult head and tail shows first 5 values
```

```python
# usage of loc [NAME OF COLUMNS-> with row and column selection]

# only radius and diganosis
# : symbol represent that select all the values in that position!

# data [ :, : ] -> select all rows and all columns

df = data.loc[ : , ['radius_mean', 'diagnosis'] ] # selects all rows as we provided : in row selection
df.tail(5)
```

```python
# : 5 -> from beginning till 5
df = data.loc[ :5 , ['radius_mean', 'diagnosis'] ] # selects all rows as we provided : in row selection
df.tail(10)
```

```python
# 565 :  -> from 565 till end
df = data.loc[ 565: , ['radius_mean', 'diagnosis'] ] # selects all rows as we provided : in row selection
df.tail(10)
```

```python
# 190 :195  -> from 190 till 195
df = data.loc[ 190:195 , ['radius_mean', 'diagnosis'] ] # selects all rows as we provided : in row selection
df.tail(10)
```

```python
data.columns
```

```python
df = data.loc[ : , 'radius_mean': 'compactness_mean']
df.head()
```

```python
# iloc
# index instead of column names

# first 5 rows and first 5 columns
df = data.iloc[ :5 , :5]
df.head()

# PYTHON HAS A LOT OF INCONSISTENCIES -> keep it only for experimentation never for production
# for prod -> use C#, Java, C++, NodeJS, anything except PYTHON!!!!!
```

```python
df = data.iloc[ 5:10 , 2:5]
df.head()
# x : y
# in loc -> x and y are both inlcuded
# in iloc-> x is included, and y is excluded
```

```python
# negative indexing is possible

df = data.iloc[ -5: , -10:-5]
# last 5 rows, and from last 10th till last 5th column
df.head()
```

```python
df = data.iloc[ 2:-2 , 2:-2]
# all rows and all columns except the first and last 2
df.head()
```
